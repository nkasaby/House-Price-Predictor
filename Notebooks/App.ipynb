```python
import tkinter
import tkinter as tk
from tkinter import PhotoImage
from tkinter import messagebox
from tkinter import OptionMenu
from tkinter import StringVar
from tkinter import ttk
from decimal import Decimal
import requests
import pandas as pd
```

```python

CS_df = pd.read_csv("../csv/city_state.csv")
CS_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
def End():
    Main.destroy()
```

```python
def Back_Main():
    Main.deiconify()
    if c == 2:
        Signin.withdraw()
    else:
        Signup.withdraw()
```

```python
def Back_Menus():
    Menus.deiconify()
    if k == 1:
        if z == 1:
            Pred_View.withdraw()
        else:
            New_Pred.withdraw()
    else:
        ViewPast.withdraw()
```

```python
def Start():
    global Main
    Main = tk.Tk()
    Main.title("Home_Scout")
    Main.geometry("600x350")
        
    label = tk.Label(Main, text = "Home Scout", font = ('Arial', 28))
    label.place(x=225, y=10)

    button_frame = tk.Frame(Main)
    button_frame.columnconfigure(0, weight=1)

    btn1 = tk.Button(button_frame, text="Sign in", font=('Arial', 22), command=Sign_In_Page)
    btn2 = tk.Button(button_frame, text="Sign up", font=('Arial', 22), command=Sign_Up_Page)

    btn1.grid(row = 0, column = 0, sticky = "news")
    btn2.grid(row = 1, column = 0, sticky = "news")

    button_frame.place(x=245, y=140)
    Main.mainloop()
```

```python
def Sign_Up_Page():
    global c
    c = 1
    global Signup, USN
    Signup = tk.Toplevel()
    Signup.title("Home_Scout")
    Signup.geometry("600x350")
    Main.withdraw()
    
    labelS = tk.Label(Signup, text="Sign Up Page", font=('Arial', 24))
    labelS.pack(pady=20)

    labelS2 = tk.Label(Signup, text="Please Enter Your New Username", font=('Arial', 24))
    labelS2.pack(pady=40)

    USN = tk.Entry(Signup, bd=10)
    USN.place(x=195, y=160)
    
    button_frame = tk.Frame(Signup)
    button_frame.columnconfigure(0, weight=1)

    btn1 = tk.Button(button_frame, text="Enter", font=('Arial', 22), command=Sign_Up)
    btn2 = tk.Button(button_frame, text="Back", font=('Arial', 22), command=Back_Main)

    btn1.grid(row = 0, column = 0, sticky = "news")
    btn2.grid(row = 1, column = 0, sticky = "news")

    button_frame.place(x=250, y=242)
```

```python
def Sign_In_Page():
    global Signin
    global c
    c = 2
    Signin = tk.Toplevel()
    Signin.title("Home_Scout")
    Signin.geometry("600x350")
    Main.withdraw()
    
    label = tk.Label(Signin, text="Sign in Page", font=('Arial', 24))
    label.pack(pady=20)

    label = tk.Label(Signin, text="Please Enter Your Username", font=('Arial', 24))
    label.pack(pady=20)

    global input
    input = tk.Entry(Signin, bd=10)
    input.place(x=195,y=160)
    
    button_frame = tk.Frame(Signin)
    button_frame.columnconfigure(0, weight=1)

    btn1 = tk.Button(button_frame, text="Enter", font=('Arial', 22), command=Sign_In)
    btn2 = tk.Button(button_frame, text="Back", font=('Arial', 22), command=Back_Main)

    btn1.grid(row = 0, column = 0, sticky = "news")
    btn2.grid(row = 1, column = 0, sticky = "news")
    button_frame.place(x=250, y=242)
```

```python
def Validate_User():
    response = requests.get("http://localhost:5000/Get_Users")
    if response.status_code == 200:
        Users = response.json()
    else:
        print("Error:", response.status_code)
    return Users

def Sign_In():
    global Username
    Username = input.get()
    if Username not in Validate_User():
        messagebox.showerror("Error", "Username Not Found, Please Try Again")
    else:
        Signin.withdraw()
        Menu()
```

```python
def Sign_Up():
    name = USN.get()
    
    if name in Validate_User() or len(name)>50:
        messagebox.showerror("Error", "Username Not Eligible, Please Try Again")

    else:
        response = requests.post("http://localhost:5000/Add_User", json=name)
        if response.status_code == 200:
            messagebox.showinfo("", "Successfully Signed Up, Please Sign In to Continue")
        else:
            print("Error:", response.status_code)
        
        Main.deiconify()
        Signup.withdraw()
```

```python
def Menu():
    global Menus
    Menus = tk.Toplevel()
    Menus.title("Home_Scout")
    Menus.geometry("600x350")
    
    label = tk.Label(Menus, text="Welcome, What do you want to do?", font=('Arial', 24))
    label.pack(pady=20)

    button_f = tk.Frame(Menus)
    button_f.columnconfigure(0, weight=1)

    btn1 = tk.Button(button_f, text="New Price Prediction", font=('Arial', 22), command=New_Predict)
    btn2 = tk.Button(button_f, text="View Past Predictions", font=('Arial', 22), command=View_Past)
    btn3 = tk.Button(button_f, text="Exit", font=('Arial', 22), command=End)

    btn1.grid(row=0, column=0, sticky="news")
    btn2.grid(row=1, column=0, sticky="news")
    btn3.grid(row=2, column=0, sticky="news")
    
    button_f.place(x=180, y=140)
```

```python
def New_Predict():
    global k 
    k = 1
    global z
    z = 2
    global New_Pred
    New_Pred = tk.Toplevel()
    New_Pred.title("Home_Scout")
    New_Pred.geometry("650x680")
    Menus.withdraw()
    
    labelS = tk.Label(New_Pred, text="Please Enter Home Info", font=('Arial', 24))
    labelS.pack(pady=20)
    
    tk.Label(New_Pred, text="Status", font=('Arial', 20)).place(x=120,y=100)
    tk.Label(New_Pred, text="Sold", font=('Arial', 20)).place(x=120,y=150)
    tk.Label(New_Pred, text="State", font=('Arial', 20)).place(x=120,y=200)
    tk.Label(New_Pred, text="City", font=('Arial', 20)).place(x=120,y=250)
    tk.Label(New_Pred, text="Zip Code", font=('Arial', 20)).place(x=120,y=300)
    tk.Label(New_Pred, text="Bedrooms", font=('Arial', 20)).place(x=120, y=355)
    tk.Label(New_Pred, text="Bathrooms", font=('Arial', 20)).place(x=120, y=405)
    tk.Label(New_Pred, text="House Size", font=('Arial', 20)).place(x=120, y=455)
    tk.Label(New_Pred, text="Acre Lot", font=('Arial', 20)).place(x=120, y=505)
    
    global clicked1, clicked2, clicked3, clicked4, clicked5, BA, BD, HS, AL
    
    clicked1 = StringVar()
    clicked1.set("None")
    
    clicked2 = StringVar()
    clicked2.set("None")
    
    clicked3 = StringVar()
    clicked3.set("None")
    
    clicked4 = StringVar()
    clicked4.set("None")
    
    clicked5 = StringVar()
    clicked5.set("None")
    
    status_list = ["for_sale", "ready_to_build"] 
    sold_list = ["yes", "no"]
    zipcode_list = list(range(1, 100))
    city_list = CS_df['city'].sort_values().unique()
    state_list = CS_df['state'].sort_values().unique()
    
    OptionMenu(New_Pred, clicked1, *status_list).place(x=360, y=108)
    OptionMenu(New_Pred, clicked2, *sold_list).place(x=360, y=158)
    OptionMenu(New_Pred, clicked3, *state_list).place(x=360, y=208)
    OptionMenu(New_Pred, clicked4, *city_list).place(x=360, y=258)
    OptionMenu(New_Pred, clicked5, *zipcode_list).place(x=360, y=308)

    BD = tk.Entry(New_Pred, bd=10)
    BD.place(x=300, y=350)
    
    BA = tk.Entry(New_Pred, bd=10)
    BA.place(x=300, y=400)
    
    HS = tk.Entry(New_Pred, bd=10)
    HS.place(x=300, y=450)
    
    AL = tk.Entry(New_Pred, bd=10)
    AL.place(x=300, y=500)
    
    button_f = tk.Frame(New_Pred)
    button_f.columnconfigure(0, weight=1)
    
    btn1 = tk.Button(button_f, text="Enter", font=('Arial', 22), command=Validate)
    btn2 = tk.Button(button_f, text="Back", font=('Arial', 22), command=Back_Menus)
    
    btn1.grid(row=0, column=0, sticky="news")
    btn2.grid(row=1, column=0, sticky="news")
    button_f.place(x=280, y=580)
    
def Validate():
    status = clicked1.get() 
    sold = clicked2.get()
    zipcode = clicked5.get()
    city = clicked4.get()
    state = clicked3.get()
    bed = BD.get()
    bath = BA.get()
    house = HS.get()
    acre = AL.get()

    if all(x != "None" for x in [status, sold, zipcode, city, state]) and all(field != "" for field in [bed, bath, house, acre]):
        Predict()
    else:
        messagebox.showinfo("Error", "Please select a value for all fields")
    
```

```python
def Validate_Pred():
    response = requests.get("http://localhost:5000/Get_Predictions")
    if response.status_code == 200:
        Preds = response.json()
    else:
        print("Error:", response.status_code)
    return Preds

def Predict():
    status = clicked1.get() 
    sold = clicked2.get()
    zip_code = clicked5.get()
    city = clicked4.get()
    state = clicked3.get()
    bed = BD.get()
    bath = BA.get()
    house_size = HS.get()
    acre_lot = AL.get()

    user_input = {
        "status": status,
        "bed": bed,
        "bath": bath,
        "acre_lot": acre_lot,
        "city": city,
        "state": state,
        "zip_code": zip_code,
        "house_size": house_size,
        "sold": sold
    }
    
    user_input = pd.DataFrame(user_input, index=[0])
    global label
    
    response = requests.post("http://localhost:5000/predict", json=user_input.to_dict(orient='records'))
    if response.status_code == 200:
        res = response.json()
        label = res[0]
    else:
        print("Error:", response.status_code) 
        
    val = (Username, status, bed, bath, acre_lot, city, state, zip_code, house_size, sold, label) 
    LVal = list(val)
    LVal[2] = int(LVal[2])  
    LVal[3] = int(LVal[3])  
    LVal[4] = Decimal(LVal[4])
    LVal[8] = Decimal(LVal[8])
    LVal = tuple(LVal)

    if LVal in Validate_Pred():
        messagebox.showerror("Error", "Already Made this Prediction")
    else:
        response = requests.post("http://localhost:5000/Add_Prediction", json=val)
        if response.status_code == 200:
            print("Prediction added to database")
        else:
            print("Error:", response.status_code)
        
        global z
        z = 1
        global Pred_View
        Pred_View = tk.Toplevel()
        Pred_View.title("Home_Scout")
        Pred_View.geometry("600x400")
        New_Pred.withdraw()
        
        labelS = tk.Label(Pred_View, text="Predicted Price Range", font=('Arial', 24))
        labelS.pack(pady=20)
        
        labelP = tk.Label(Pred_View, text=label, font=('Arial', 24), foreground="green")
        labelP.pack(pady=20)
        
        labelC = tk.Label(Pred_View, text="Please Provide Your Feedback", font=('Arial', 24))
        labelC.pack(pady=20)
        
        
        tk.Button(Pred_View, text="Correct", font=('Arial', 22), command=Correct).place(x=300, y=250)
        tk.Button(Pred_View, text="Report", font=('Arial', 22), command=Report).place(x=200, y=250)
        tk.Button(Pred_View, text="Back", font=('Arial', 22), command=Back_Menus).place(x=260, y=300)
    
def Correct():
    status = clicked1.get() 
    sold = clicked2.get()
    zipcode = clicked5.get()
    city = clicked4.get()
    state = clicked3.get()
    bed = BD.get()
    bath = BA.get()
    house = HS.get()
    acre = AL.get()
    
    Val = (status, bed, bath, acre, city, state, zipcode, house, sold, label)
    response = requests.post("http://localhost:5000/Add_Train", json=Val)
    if response.status_code == 200:
        print("Point added to train")
    else:
        print("Error:", response.status_code)
    
    Pred_View.withdraw()
    Back_Menus()
    messagebox.showinfo("", "Thank You for your Feedback")
    
def Report():
    zipcode = clicked5.get()
    city = clicked4.get()
    state = clicked3.get()
    bed = BD.get()
    bath = BA.get()
    house = HS.get()
    acre = AL.get()
    
    Val = (Username, bed, bath, house, acre, city, state, zipcode, label)
    
    response = requests.post("http://localhost:5000/Add_Reported", json=Val)
    if response.status_code == 200:
        print("Point added to train")
    else:
        print("Error:", response.status_code)
    
    Feedback()
    
```

```python
def Get_Values():
    price = clicked.get()
    if price == "Price Ranges":
        messagebox.showinfo("Error", "Please select a Price Range")
    elif price == label:
        messagebox.showinfo("", "The Selected Price Range is the Predicted")
    else:
        Feed_Back.withdraw()
        Back_Menus()
        messagebox.showinfo("", "Reported Successfully")
        
        status = clicked1.get() 
        sold = clicked2.get()
        zipcode = clicked5.get()
        city = clicked4.get()
        state = clicked3.get()
        bed = BD.get()
        bath = BA.get()
        house = HS.get()
        acre = AL.get()
        
        Val = (status, bed, bath, acre, city, state, zipcode, house, sold, price)
        
        response = requests.post("http://localhost:5000/Add_Train", json=Val)
        if response.status_code == 200:
            print("Point added to train")
        else:
            print("Error:", response.status_code)


def Feedback():
    global Feed_Back
    Feed_Back = tk.Toplevel()
    Feed_Back.title("Home_Scout")
    Feed_Back.geometry("600x350")
    Pred_View.withdraw()
    
    labels = tk.Label(Feed_Back, text="Please Choose the Correct Price Range", font=('Arial', 24))
    labels.pack(pady=20)
    
    global clicked
    clicked = StringVar()
    clicked.set("Price Ranges")

    price_list = ['1k - 89k', '90k - 187k', '188k - 373k', '374k - 699k', '700k - 1424k', '1425k - 875000k']

    OptionMenu(Feed_Back, clicked, *price_list).place(x=240, y=100)
    tk.Button(Feed_Back, text="Enter", font=('Arial', 22), command=Get_Values).place(x=255, y=250)
    
```

```python
def Validate_Report():
    response = requests.get("http://localhost:5000/Get_Reports")
    if response.status_code == 200:
        Reports = response.json()
    else:
        print("Error:", response.status_code)
    return Reports

def View_Past():
    global k
    k = 2
    global ViewPast
    ViewPast = tk.Toplevel()
    ViewPast.title("Home_Scout")
    ViewPast.geometry("800x350")
    Menus.withdraw()
        
    response = requests.get("http://localhost:5000/View_Past_Predictions", json=Username)
    if response.status_code == 200:
        values = response.json()
    else:
        print("Error:", response.status_code)
    
    if len(values) == 0:
        messagebox.showerror("", "No Past Predictions Found")
        Back_Menus()
    else:
        def Report_Selected():
            selected_item = tree.focus()
            if selected_item == '':
                messagebox.showerror("Error", "Please select a row to report!")
                return
            else:
                selected_data = tree.item(selected_item, 'values')
                Val = (Username,) + selected_data
                
                LVal = list(Val)
                LVal[1] = int(LVal[1])  
                LVal[2] = int(LVal[2])  
                LVal[3] = str(LVal[3])
                LVal[4] = str(LVal[4])
                
                print(LVal)
                print(Validate_Report())
                
                if LVal in Validate_Report():
                    messagebox.showerror("Error", "Already Reported this Prediction")

                else:
                    response = requests.post("http://localhost:5000/Add_Reported", json=Val)
                    if response.status_code == 200:
                        print("Point Reported")
                    else:
                        print("Error:", response.status_code)
  
        tree = ttk.Treeview(ViewPast, selectmode='browse')
        tree["columns"] = ("bed", "bath", "size", "acre", "city", "state", "zipcode", "price")
            
        tree.column("#0", width=0, minwidth=0)
        tree.column("bed", width=100, minwidth=100, anchor='center')
        tree.column("bath", width=100, minwidth=100, anchor='center')
        tree.column("size", width=100, minwidth=100, anchor='center')
        tree.column("acre", width=100, minwidth=100, anchor='center')
        tree.column("city", width=100, minwidth=100, anchor='center')
        tree.column("state", width=100, minwidth=100, anchor='center')
        tree.column("zipcode", width=100, minwidth=100, anchor='center')
        tree.column("price", width=100, minwidth=100, anchor='center')
        
        tree.heading("bed", text="Beds", anchor='center')
        tree.heading("bath", text="Baths", anchor='center')
        tree.heading("size", text="House_Size", anchor='center')
        tree.heading("acre", text="Acre_Lot", anchor='center')
        tree.heading("city", text="City", anchor='center')
        tree.heading("state", text="State", anchor='center')
        tree.heading("zipcode", text="Zip_Code", anchor='center')
        tree.heading("price", text="Price_Range", anchor='center')
            
        i = 0
        for j in values:
            tree.insert('', i, text="", values=(j[0], j[1], j[2], j[3], j[4], j[5], j[6], j[7]))
            i = i + 1
        tree.pack()
        
        button_f = tk.Frame(ViewPast)
        button_f.columnconfigure(0, weight=1)
        
        btn1 = tk.Button(button_f, text="Report", font=('Arial', 22), command=Report_Selected)
        btn2 = tk.Button(button_f, text="Back", font=('Arial', 22), command=Back_Menus)
        
        btn1.grid(row=0, column=0, sticky="news")
        btn2.grid(row=1, column=0, sticky="news")
        button_f.place(x=350, y=240)
        
        
```
